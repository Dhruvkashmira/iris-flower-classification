#PROJECT IRIS FLOWER CLASSIFICATION

##Discription

The aim of this project is to classify Iris flowers into three species—Setosa, Versicolor, and Virginica—based on their physical characteristics.
Using a machine learning models, we analyze four key features of the Iris dataset:

Sepal Length

Sepal Width

Petal Length

Petal Width

The model learns patterns from these features and accurately predicts the flower species.

```python
#importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#Importing the Dataset


```python
columns =['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']
#load data
df = pd.read_csv('/content/Iris.csv',names=columns)
df.head(150)
```

#Data Visualization


```python
df.describe()
```

```python
#WHOLE DATASET VISUALIZATION
sns.pairplot(df,hue='Species')
```

SEPRATING INPUT AND OUTPUT COLUMNS


```python
data =df.values
x = data[:,0:4]
y = data[:,4]
print(x)
print(y)
```

FOR TESTING AND TRAINING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
#pridiction
pridiction = model_LR.predict(x_test)
#calculating pridiction
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pridiction)*100)

for  i in range(len(pridiction)):
  print(pridiction[i],y_test[i])
```

#svg model


```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(x_train,y_train)
```

```python
#pridection by svc model
pridiction2 = model_svc.predict(x_test)
#calculating pridiction
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pridiction2)*100)
```

decicion tree algo


```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
```

```python
pridection3 = model_DTC.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pridection3)*100)
```

```python
check_new = np.array([[4,5,2.4,0.7],[4.8,0.3,4.2,1.2],[3.3,5.5,4.6,1.9]])
pridiction = model_LR.predict(check_new)
print("Prediction of species: {}" .format(pridiction))
```
